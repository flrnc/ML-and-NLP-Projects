```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("age.csv")
df.head(7)
```

```python
sd = df.sort_values(['Age'])
bar = sd.plot(kind = 'bar', x = 'Age')
bar.set_title('AGE VS TIME (Bar Graph)')
bar.set_xlabel('Age')
bar.set_ylabel("Time Spent in Hours")
```

```python
ax1 = sd.plot.scatter(x = 'Age', y = 'Watching Television', label = 'Watching Television', color = 'blue')    
ax2 = sd.plot.scatter(x = 'Age', y = 'Using Social Media', label = 'Using Social Media', color = 'orange', ax = ax1)    
ax3 = sd.plot.scatter(x = 'Age', y = 'Taking Online Class', label = 'Using Social Media', color = 'green', ax = ax1)
ax4 = sd.plot.scatter(x = 'Age', y = 'Taking Modular Class', label = 'Taking Modular Class', color = 'red', ax = ax1)

ax1.set_title('AGE VS TIME ( Scatter Map )')
```

```python
hbar = sd.plot.barh(x = 'Age')
hbar.set_title('AGE VS TIME (Barh Graph)')
hbar.set_xlabel('Age')
hbar.set_ylabel('Time Spent in Hours')
```

```python
line = sd.plot.line(x = 'Age')
line.set_title('AGE VS TIME (Line Graph)')
line.set_xlabel('Age')
line.set_ylabel('Time Spent in Hours')
```

```python
area = sd.plot.area(x = 'Age')
area.set_title('AGE VS TIME (Area Graph)')
area.set_xlabel('Age')
area.set_ylabel('Time Spent in Hours')
```
